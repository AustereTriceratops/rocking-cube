```python
hashTable = {} # permutation hash : move sequence from identity element

for i in range(1, 13):
  generated_moves = generate_all_moves_of_len(i, oriented=True)

  for sequence in generated_moves:
      permutation = parseMoves(sequence)
      hashed_permutation = hash(tuple(permutation))
      
      if hashed_permutation in hashTable:
        hashTable[hashed_permutation].append(sequence)
      else:
        hashTable[hashed_permutation] = [sequence]

for value in hashTable.values():
  if len(value) > 1:
    print(value)
```

```python
unique_oriented_configurations = set()

# find all unique orientations of the cube
for i in range(17):
  generated_moves = generate_all_moves_of_len(i, oriented=True)
  
  for sequence in generated_moves:
    permutation = parseMoves(sequence)
    unique_oriented_configurations.add(tuple(permutation))
  
  print(len(unique_oriented_configurations))
```

```python
n_cycle_count = {}
unique_permutations = set()

# count the number of 2-cycles, 3-cycles, etc.
for i in range(17):
  generated_moves = generate_all_moves_of_len(i, oriented=True)
  
  for sequence in generated_moves:
    order = permutationOrder(sequence)
    permutation = parseMoves(sequence)
    
    if not tuple(permutation) in unique_permutations:
      unique_permutations.add(tuple(permutation))
      if order in n_cycle_count:
        n_cycle_count[order] += 1
      else:
        n_cycle_count[order] = 1
  
print(n_cycle_count)
```
